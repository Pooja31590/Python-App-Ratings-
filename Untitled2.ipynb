```python
#1. find the name of all the countries where land area is > 20000 sq km without printing
  # index column in the o/p
import pandas
df1=pandas.read_csv('C:\\Users\\pooja\\Downloads\\AllCountries.csv')
df1.head()
```

```python
for col in df1.itertuples():
    if col[2]>2000:
        print(col[1])
```

```python
#2. find the relation between GDP and birth rate using scatter plot
import matplotlib.pyplot as plt
x=df1.BirthRate
y=df1.GDP
plt.scatter(x,y)
plt.xlabel('BirthRate')
plt.ylabel('GDP')
plt.show()
```

```python
#3. make a pie chart of top 10 countries having highest GDP
df_sorted=df1.sort_values(by='GDP', ascending= False).head(10)
df_sorted
```

```python
plt.pie(df_sorted.GDP,labels=df_sorted.Country)
plt.show()
```

```python
#4. count the missing values in each column and print in ascending order
df2=df1.isnull().sum()
df2_sorted=df2.sort_values(ascending=True)
df2_sorted
```

```python
#5. How GDP is related to Death Rate. plot scatter plot
x=df1.GDP
y=df1.DeathRate
plt.scatter(x,y)
plt.xlabel('GDP')
plt.ylabel('DeathRate')
plt.show()
```

```python
#6. List down top 5 countries where Female Labor is maximum and compare their GDPs in
#  a form of bar chart
df1_fL=df1.sort_values(by='FemaleLabor', ascending=False).head(5)
df1_fL
```

```python
d={}
for col in df1_fL.itertuples():
    d[col[1]]=col[4]
    print(d)
```

```python
import numpy
for i, k in enumerate(d):
    plt.bar(i,d[k])
plt.xticks(numpy.arange(len(d)), d.keys())
plt.show()
```
